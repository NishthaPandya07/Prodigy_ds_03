```python
plt.figure(figsize=(15,8))
sentiment_by_topic = data.groupby(['Topic', 'Sentiment']).size().unstack(fill_value=0)
sentiment_by_topic.plot(kind='bar', stacked=True)
plt.title('Sentiment by Topic')
plt.xlabel('Topic')
plt.ylabel('Count')
plt.show()
```

```python
senti_count = data['Sentiment'].value_counts()
senti_count
plt.figure(figsize=(6, 3))
senti_count.plot(kind='bar', color=['blue', 'red', 'yellow', 'green'])
plt.title('Sentiment Distribution')
plt.ylabel('Number of Tweets')
plt.xlabel('Sentiment')
plt.show()
```

```python
data_brand = data[data['Topic'].str.contains('Microsoft', case=False)]
data_brand_count = data_brand['Sentiment'].value_counts()
data_brand_count
plt.figure(figsize=(6, 6))
plt.pie(data_brand_count, labels=data_brand_count.index, startangle=140, autopct='%1.1f%%')
plt.title('Sentiment Distribution for Microsoft')
plt.show()
```

```python
data.Topic.unique()
Topic_count=data['Topic'].value_counts()
top_10_Topic = Topic_count.head(10)
plt.figure(figsize=(10, 6))
top_10_Topic.plot(kind='bar')
plt.xlabel("Topics")
plt.ylabel('Tweet Count')
plt.title('Top 10 Topics with Highest Tweet Engagement')
plt.show()
```

```python
plt.figure(figsize=(15, 7))
sns.countplot(data=data, y=data.columns[1], order=data[data.columns[1]].value_counts().index, palette='viridis')
plt.title('Topic Distribution in Training Data')
plt.xlabel('Count')
plt.ylabel('Topic')
plt.tight_layout()
plt.show()
```
